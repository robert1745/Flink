```python
import numpy as np
import pandas as pd
garbage_value = -2147483650

# Optimized forward-fill function
def ffill_fxn_optimized(df, garbage_value):
    # Identify columns starting with 'BF_'
    bf_columns = [col for col in df.columns if col.startswith('BF_')]
    
    # Replace garbage values with NaN for easier handling
    df[bf_columns] = df[bf_columns].replace(garbage_value, np.nan)
    
    # Use Pandas' built-in forward-fill method
    df[bf_columns] = df[bf_columns].ffill(axis=0)
    
    return df

# Apply the optimized function
df = ffill_fxn_optimized(df, garbage_value)
print(df)
```
